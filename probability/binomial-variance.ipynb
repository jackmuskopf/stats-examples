```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def variance(p, n):
    return p*(1-p)*n
```

```python
Nn, Np = 1000, 1000
n0 = 4

ns = list(range(n0, Nn+n0))
ps = np.linspace(0, 1, Np)
res = np.zeros(shape=(Nn, Np))
for i, n in enumerate(ns):
    for j, p in enumerate(ps):
        res[i,j] = variance(p, n)
```

```python
df = pd.DataFrame(res.T, columns=ns, index=pd.Series(np.round(ps, 2)))
```

```python
plt.figure(figsize=(16,13))
sns.heatmap(df)
```

```python
# when does binomial distribution have variance of 1?
# var(binomial(n, p)) = p(1-p)*n = 1 => np^2 - np + 1 = 0
n = 30
p, q = np.roots((n, -n, 1))
if np.isclose(variance(p, n), 1):
    print(f'Variance of binomial where p={p} and n={n} is close to 1.')
else:
    print(f'Failed to find p for n={n} with binomial variance of 1.')
```
