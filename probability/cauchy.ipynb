```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# default seaborn kwargs
w, h = (15, 8)
sns.set(rc = {'figure.figsize':(w, h)})
_sns = dict(height=h, aspect=w/h)
```

```python
x = np.linspace(-np.pi/2, np.pi/2, 1000)
y = np.tan(x)
sns.lineplot(x=x, y=y)
plt.ylim(-50, 50)
```

Consider $ X \sim U(-\pi/2, \pi/2) $ ; let $ Y = tan(X) $. Then,

\begin{align}F_Y(y) &= P(Y\le y) \\&= P(\tan(X)\le y) \\&= P(X\le \tan^{-1}(y))\\&=F_X(\tan^{-1}y) \\&=  \frac{1}{\pi}\left({\tan^{-1}y}+\frac{\pi}2 \right)  \\& \implies f_Y(y) =  \frac{1}{\pi} \frac{1}{1+y^2} \end{align}

```python
py = np.vectorize(lambda y: (1/np.pi)*(1/(1+y**2)))
x = np.linspace(-10, 10, 1000)
y = py(x)
```

```python
sns.lineplot(x=x, y=y)
```

This is a cauchy distribution with $ x_0 = 0 $ and $ \gamma = 1 $
