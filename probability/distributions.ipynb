### [Normal distribution](https://en.wikipedia.org/wiki/Normal_distribution)
Parameters:
- $\mu \in \mathbb{R}$ ~ location (mean)
- $\sigma^2 > 0$ ~ scale (variance) <br>

Support: $ x \in \mathbb{R} $ <br>
Mean: $\mu$ <br>
Median: $\mu$ <br>
Mode: $\mu$ <br>
Variance: $\sigma^2$ <br>
PDF: $\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}$ <br>
CDF: $ \frac{1}{2}[1+ $ erf $(\frac{x-\mu}{\sigma\sqrt{2}})]$

Frequently occuring in various applications, often as a consequence of the [central limit theorem](https://en.wikipedia.org/wiki/Central_limit_theorem).

```python
from scipy.stats import norm as normal
```

```python
mu = 0
sigma = 0.5
```

```python
rv = normal(loc=mu, scale=sigma)
mean, var, skew, kurtosis = rv.stats(moments='mvsk')
mean, var, skew, kurtosis
```

```python
pdf, cdf = rv.pdf, rv.cdf
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)

fig, ax = plt.subplots(1, 1)
ax.plot(x, pdf(x), 'k-', lw=2, label='PDF')
ax.plot(x, cdf(x), 'r-', lw=2, label='CDF')
ax.legend(loc='best', frameon=False)
plt.show()
```

### [Log-normal distribution](https://en.wikipedia.org/wiki/Log-normal_distribution)

Parameters:
- $\mu \in \mathbb{R}$ ~ scale
- $\sigma > 0$ ~ shape <br>

Support: $x \in (0, \infty) $ <br>
Mean: $ e^{(\mu + \frac{\sigma^2}{2})}$ <br>
Median: $ e^\mu $ <br>
Mode: $ e^{\mu - \sigma^2} $ <br>
Variance: $(e^{\sigma^2} - 1)(e^{2\mu + \sigma^2})$ <br>
PDF: $\frac{1}{x\sigma\sqrt{2\pi}}e^{-\frac{1}{2}(\frac{ln(x)-\mu}{\sigma})^2}$ <br>
CDF: $ \frac{1}{2}[1+ $ erf $(\frac{ln(x)-\mu}{\sigma\sqrt{2}})]$

A distribution for random variables whose logarithm is normally distributed.  If $X$ follows a normal distribution, then $e^X$ follows a log-normal distribution.

[Applications](https://en.wikipedia.org/wiki/Log-normal_distribution#Occurrence_and_applications)
- [The length of comments posted in Internet discussion forums follows a log-normal distribution](https://en.wikipedia.org/wiki/Log-normal_distribution#cite_note-46)
- [Users' dwell time on online articles (jokes, news etc.) follows a log-normal distribution](https://en.wikipedia.org/wiki/Log-normal_distribution#cite_note-47)

```python
from scipy.stats import lognorm
```

```python
sigma = 1
mu = 0
rv = lognorm(scale=np.exp(mu), s=sigma)
```

```python
mean, var, skew, kurt = rv.stats(moments='mvsk')
mean.item(), var.item(), skew.item(), kurt.item()
```

```python
np.log(mean)
```

```python
pdf, cdf = rv.pdf, rv.cdf
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)

fig, ax = plt.subplots(1, 1)
ax.plot(x, pdf(x), 'k-', lw=2, label='PDF')
ax.plot(x, cdf(x), 'r-', lw=2, label='CDF')
ax.legend(loc='best', frameon=False)
plt.show()
```

### [Chi Squared distribution](https://en.wikipedia.org/wiki/Chi-squared_distribution)

The sum of squares $k$ normally distributed random variables follows a chi-squared distribution with $k$ degrees of freedom.

Parameters:
- $ k \in \mathbb{N} $ ~ degrees of freedom

Support: $x \in (0, \infty) $ if $k=1$ else $x \in [0, \infty]$ <br>
Mean: $ k $ <br>
Median: $ k(1-\frac{2}{9k})^3 $ <br>
Mode: $ max(k-2, 0) $ <br>
Variance: $ 2k $ <br>
PDF: $ \frac{1}{2^{k/2}\Gamma(k/2)}x^{\frac{k}{2}-1}e^{-x/2} $ <br>
CDF: $ \frac{1}{\Gamma(k/2)}\gamma(k/2, x/2) $

[Applications](https://en.wikipedia.org/wiki/Chi-squared_distribution#Occurrence_and_applications)
- Sum of squares of normally distributed variables
- [MRI](https://linkinghub.elsevier.com/retrieve/pii/S1120179714000829)
- Estimating variances; goodness of fit
- [Pearson's Chi-squared test](https://en.wikipedia.org/wiki/Pearson%27s_chi-squared_test)

### [Student's T-distribution](https://en.wikipedia.org/wiki/Student's_t-distribution)

If $\{ X_1, X_2, ... X_n \} $ are i.i.d normally distributed random variables with mean $ \mu $, then $ \frac{\bar{X}-\mu}{S/\sqrt{n}} $ follows a T-distribution with $n$ degrees of freedom.

Parameters:
- $ \nu > 0 $ ~ degrees of freedom

Support: $x \in \mathbb{R} $ <br>
Mean: $ 0 $ if $ \nu > 0 $ else undefined <br>
Median: $ \nu $ <br>
Mode: $ \nu $ <br>
Variance: $ \frac{\nu}{\nu - 2} $ if $ \nu > 2 $ else $ \infty $ if $ x \in (1, 2] $ else undefined <br>
PDF: [see](https://en.wikipedia.org/wiki/Student's_t-distribution) <br>
CDF: [See](https://en.wikipedia.org/wiki/Student's_t-distribution)

[Applications](https://en.wikipedia.org/wiki/Student's_t-distribution#Uses)
- T-test
- Confidence intervals
- [Prediction intervals](https://en.wikipedia.org/wiki/Prediction_interval)

### [F-distribution](https://en.wikipedia.org/wiki/F-distribution)

If X,Y are independent random variables following chi-squared distributions with $d_x, d_y$ degrees of freedom, then $ \frac{X/d_x}{Y/d_y} \sim F(d_x, d_y). $

Parameters:
- $ d_1, d_2 > 0 $ ~ degrees of freedom


[Applications](https://en.wikipedia.org/wiki/F-distribution#Properties_and_related_distributions)
- [ANOVA and F-test](https://en.wikipedia.org/wiki/F-test)


```python
from scipy.stats import f
```

### [Pareto distribution](https://en.wikipedia.org/wiki/Pareto_distribution)

Parameters:
- $ x_m > 0 $ ~ scale
- $ \alpha > 0 $ ~ shape <br>

Support: $x \in [x_m, \infty) $ <br>
Mean: $ \frac{\alpha x_m}{\alpha - 1}$ if $ \alpha > 1 $ else $ \infty $ <br>
Median: $ x_m \sqrt[\alpha]{2} $ <br>
Mode: $ x_m $ <br>
Variance: $ \frac{x_m^2}{(\alpha-1)^2(\alpha-2)}  $ if $ \alpha > 2 $ else $ \infty $ <br>
PDF: $ \frac{\alpha x_m^\alpha}{x^{\alpha + 1}} $ <br>
CDF: $ 1 - (\frac{x_m}{x})^\alpha $

[Applications](https://en.wikipedia.org/wiki/Pareto_distribution#Occurrence_and_applications)
- Size of human settlements
- Size of sand particles
- Length of hard drive error burst


```python
from scipy.stats import pareto
```

```python
params = dict(
    b=1.5, # alpha
    loc=10,
    scale=10
) 
rv = pareto(**params)
```

```python
mean, var, skew, kurt = rv.stats(moments='mvsk')
mean.item(), var.item(), skew.item(), kurt.item()
```

```python
pdf, cdf = rv.pdf, rv.cdf
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax1.set_title('Pareto Distribution')
ax1.plot(x, pdf(x), 'k-', lw=2, label='PDF')
ax2.plot(x, cdf(x), 'r-', lw=2, label='CDF')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax1.legend(loc='upper left', frameon=False), ax2.legend(loc='upper right', frameon=False)
plt.show()
```

### [Continuous uniform distribution](https://en.wikipedia.org/wiki/Continuous_uniform_distribution)

Parameters:
- $ - \infty < a < b < \infty $

Support: $x \in [a, b] $ <br>
Mean: $ \frac{1}{2}(a + b) $ <br>
Median: $ \frac{1}{2}(a + b) $ <br>
Mode: $ x \in [a,b] $ <br>
Variance: $ \frac{(b-a)^2}{12} $ <br>
PDF: $ \frac{1}{b-a} $ if $ x \in [a,b] $ else 0 <br>
CDF: 0 if $ x < a $ else $ \frac{x-a}{b-a} $ if $ x \in [b,a] $ else 1

[Applications](https://en.wikipedia.org/wiki/Continuous_uniform_distribution#Occurrence_and_applications)

### [Binomial distribution](https://en.wikipedia.org/wiki/Binomial_distribution)

Parameters:
- $ n \in \mathbb{N}$
- $ p \in [0, 1] $
- $ q = 1-p $

Support: $x \in \{0, 1, ..., n\} $ <br>
Mean: $ np $ <br>
Median: $ \in[\lfloor{np}\rfloor, \lceil{np}\rceil] $ <br>
Mode: $ \lfloor (n + 1)p \rfloor $ or $ \lceil (n + 1)p \rceil - 1 $ <br>
Variance: $ npq $ <br>
PMF: $f(k) = {n \choose k} p^k q^{(n-k)} $ <br>
CDF: $F(k) = \sum_{i=0}^{\lfloor k \rfloor} f(i) $


```python
from scipy.stats import binom
```

```python
n, p = 10000, 0.01
rv = binom(n, p)
```

```python
fig, ax = plt.subplots(1, 1)
ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis

xs, xe = (binom.ppf(0.0001, n, p),
              binom.ppf(0.9999, n, p))
x = np.arange(xs, xe)
xcdf = np.linspace(xs, xe, 1000)

ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
ax.legend(loc='upper left', frameon=False)

ax2.plot(xcdf, rv.cdf(xcdf), 'k-', label='binomial cdf')
ax2.legend(loc='upper right', frameon=False)

plt.show()
```

Parameters:
- $ r \in \mathbb{N}$
- $ p \in [0, 1] $
- $ q = 1-p $

Support: $x \in \{0, 1, ..., n\} $ <br>
Mean: $ \frac{rp}{q} $ <br>
Median: ? <br>
Mode: $ \lfloor \frac{p(r-1)}{q} \rfloor $ if $ r > 1 $ else 0 <br>
Variance: $ \frac{pr}{q^2} $ <br>
PMF: $f(k) = {k+r-1 \choose k} p^k q^r $ <br>
CDF: $F(k) = \sum_{i=0}^{\lfloor k \rfloor} f(i) $

### [Geometric distribution](https://en.wikipedia.org/wiki/Geometric_distribution)

In a sequence of Bernoulli experiments, stop after $ 1 $ success.  The number of experiments that occurred follows a geometric distribution.

Parameters:
- $ p \in [0, 1] $

Support: $x \in \mathbb{N} $ <br>
Mean: $ \frac{1}{p} $ <br>
Median: $ \lceil \frac{-1}{log_2(1-p)} \rceil $ <br>
Mode: 1 <br>
Variance: $ \frac{1-p}{p^2} $ <br>
PMF: $f(k) = (1-p)^{(k-1)}p$ <br>
CDF: $ F(k) = 1 - (1-p)^k $


```python
from scipy.stats import geom
```

```python
p = 0.1
rv = geom(p=p, loc=1)
```

```python
mean, var, skew, kurt = rv.stats(moments='mvsk')
mean.item(), var.item(), skew.item(), kurt.item()
```

```python
fig, ax = plt.subplots(1, 1)
ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis

xs, xe = (rv.ppf(0.001),
              rv.ppf(0.999))
x = np.arange(xs, xe)
xcdf = np.linspace(xs, xe, 1000)

ax.plot(x, rv.pmf(x), 'bo', ms=8, label='geom pmf')
# ax.vlines(x, 0, rv.pmf(x), colors='b', lw=5, alpha=0.5)
ax.legend(loc='upper left', frameon=False)

ax2.plot(xcdf, rv.cdf(xcdf), 'k-', label='geom cdf')
ax2.legend(loc='upper right', frameon=False)

plt.show()

# the PMF is strictly decreasing; as k increases, f(k) decreases
```

### [Multinomial distribution](https://en.wikipedia.org/wiki/Multinomial_distribution)

A multinomial distribution is appropriate for a k-sided die rolled n times.

Parameters:
- $ \{p_1, p_2, ... p_k\} \subset [0, 1] $
- $ n \in \mathbb{N}$ ~ the number of trials

Support: $ x_i \in \mathbb{N}, x_i \leq n, \sum{x_i} = n $ <br>
Mean: $ E[x_i] = p_i $ <br>
Median: $ \lceil \frac{-1}{log_2(1-p)} \rceil $ <br>
Mode: 1 <br>
Variance: $ Var(X_i) = np_i(1-p_i )$ <br>
Covariance: $ Cov(X_i, X_j) = -np_ip_j, i \neq j $ <br>
PMF: $ f(\vec{x}) = \frac{n!}{\Pi_i {x_i!}} \Pi_i {{p_i}^{x_i}} $ <br>
CDF: $ P(x_1<a_1, x_2<a_2, ... x_k<a_k) = \sum_{x_i < a_i, \forall i \in \{1, 2, ... k\}}f(\vec{x}) $

```python
from scipy.stats import multinomial
```

```python
# define shape of rv
k = 80
d = 3

# define probability
p = np.random.uniform(low=0, high=1, size=d)
# p = np.ones(d)

p = p/p.sum()
rv = multinomial(k, p)
```

```python
# generate some nonzero point
_x = list()
while len(_x) < d-1:
    _x.append(np.random.randint(0, k-sum(_x)))
_x.append(k - sum(_x))
rv.pmf(np.array(_x))
```

```python
TRANSFORM_PLANE = False
```

```python
# prep test points for PMF plot
x = [np.arange(0, k) for _ in range(d)]
x1 = np.meshgrid(*x)
X = np.array([_x.ravel() for _x in x1]).T
X = np.hstack((X[:, -1].reshape(-1, 1), X[:, :-1]))
```

```python
# linear transform into plane where sum(x_i) = k; need to improve rescaling of Xh for visibility; best to leave this false for now
Xh = X.copy()
if TRANSFORM_PLANE:
    Xh = np.apply_along_axis(
        lambda _col : _col + np.ones(d)*(k-_col.sum())/d,
        1, 
        Xh
    )

    # translate for visibility of distribution
    _d = -1*np.ones(d)
    _d[-1] = d-1
    Xh += _d*5
```

```python
# generate distribution in plane
ps = np.zeros(k**d)
for _x, _xh in zip(X, Xh):
    index = 0
    for _d, x_i in enumerate(_x):
        index += x_i*(k**_d)
    ps[index] = rv.pmf(_xh) if _xh.sum() == k else 0
```

```python
# sum over axis that aren't displayed (for marginal PMF)
ps = ps.reshape(*(k for _ in range(d)))
while len(ps.shape) > 2:
    ps = ps.sum(axis=-1)
```

```python
# plot
fig, ax = plt.subplots()
ax.imshow(ps)
```

```python
SCALE_UP=False # bring probabilities closer to one to see pattern

fig, ax = plt.subplots()
cs = ax.contourf(x[0], x[1], np.power(ps, 3/k) if SCALE_UP else ps, levels=200, cmap='Spectral_r')
cbar = fig.colorbar(cs)
plt.axis('scaled')
```
